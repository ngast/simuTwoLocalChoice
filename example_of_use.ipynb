```python
import numpy as np
import matplotlib.pyplot as plt
from twochoiceneighboor import TwoChoiceNeighboor
```

This notebook demonstrates how to use the class "TwoChoiceNeighboor".

Note: the first time the simulation are run, this takes quite some times because the simulation simulates $10^9$ events. The simulation results are then saved in a file which means that this is much faster after that. 


## Case of a local choice when the geometry is a line

```python
def compare_all(rho, logscale=False):
    model = TwoChoiceNeighboor(rho)
    plt.plot(model.simulate(100),'+', label='simulation (JSQ-2 on ring)')
    plt.plot(model.simulate(100, choice=2),'+', label='simulation (classical JSQ-2, no geometry)')
    plt.plot(model.pair_approximation(), '--', label='pair approximation')
    plt.plot(model.mean_field_approximation(), ':', label='mean field approximation')
    #plt.plot(model.mm1(), 'x', label='mean field approximation')
    xmax = next((i for i,x in enumerate(model.simulate(100)) if x<1e-7))
    plt.xlim(0,max(10,xmax))
    plt.legend()
    if logscale:
        plt.semilogy([],[])
        plt.ylim([1e-7,1])
```

```python
i=1
fig = plt.figure()
fig.set_size_inches((15,10))
for rho in [0.7, 0.9, 0.95]:
    plt.subplot(3,2,i)
    compare_all(rho, logscale=False)
    i+=1
    plt.subplot(3,2,i)
    compare_all(rho, logscale=True)
    i+=1
```

## when the geometry is a 2D square.

```python
model = TwoChoiceNeighboor(0.7)
plt.plot(model.simulate(100, choice=3),'+', label='simulation (square)')
plt.plot(model.pair_approximation(degree=4), '--', label='pair approximation, degree=4')
plt.plot(model.mean_field_approximation(), '--', label='mean field approximation')
plt.legend()
plt.xlim([0,10])
```
