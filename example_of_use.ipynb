```python
import numpy as np
import matplotlib.pyplot as plt
from twochoiceneighboor import TwoChoiceNeighboor
```

This notebook demonstrates how to use the class "TwoChoiceNeighboor"

## Case with local choice (choice = line)

```python
def compare_all(rho, logscale=False):
    model = TwoChoiceNeighboor(rho)
    plt.plot(model.simulate(100),'+', label='simulation (JSQ-2 on ring)')
    plt.plot(model.simulate(100, choice=2),'+', label='simulation (classical JSQ-2, no geometry)')
    plt.plot(model.pair_approximation(), '--', label='pair approximation')
    plt.plot(model.mean_field_approximation(), ':', label='mean field approximation')
    #plt.plot(model.mm1(), 'x', label='mean field approximation')
    plt.xlim(0,10)
    plt.legend()
    if logscale:
        plt.semilogy([],[])
        plt.xlim([0,20])
        plt.ylim([1e-10,1])
```

```python
i=1
fig = plt.figure()
fig.set_size_inches((15,10))
for rho in [0.7, 0.9, 0.95]:
    plt.subplot(3,2,i)
    compare_all(rho, logscale=False)
    i+=1
    plt.subplot(3,2,i)
    compare_all(rho, logscale=True)
    i+=1
```
